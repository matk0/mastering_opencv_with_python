```python
import cv2
import numpy as np
import matplotlib.pyplot as plt 
%matplotlib inline
from IPython.display import Image
```

```python
image = 'new-zealand-lake.jpg'
img = cv2.imread(image, cv2.IMREAD_COLOR)

Image(filename=image, width=400)
```

```python
matrix = np.ones(img.shape, dtype="uint8") * 30

img_brighter = cv2.add(img, matrix)
img_darker = cv2.subtract(img, matrix)

plt.figure(figsize=[18,5])
plt.subplot(131); plt.imshow(img_darker[:, :, ::-1]);   plt.title('Darker')
plt.subplot(132); plt.imshow(img[:, :, ::-1]);          plt.title('Original')
plt.subplot(133); plt.imshow(img_brighter[:, :, ::-1]); plt.title('Brighter');
```

```python
increasing_contrast_matrix = np.ones(img.shape) * 1.2
decreasing_contrast_matrix = np.ones(img.shape) * 0.8

lower_contrast_image = np.uint8(cv2.multiply(np.float64(img), decreasing_contrast_matrix))
higher_contrast_image = np.uint8(cv2.multiply(np.float64(img), increasing_contrast_matrix))

plt.figure(figsize = [18,5])
plt.subplot(131); plt.imshow(lower_contrast_image[:, :, ::-1]);  plt.title('Lower Contrast')
plt.subplot(132); plt.imshow(img[:, :, ::-1]);        plt.title('Original')
plt.subplot(133); plt.imshow(higher_contrast_image[:, :, ::-1]); plt.title('Higher Contrast');
```

```python
increasing_contrast_matrix = np.ones(img.shape) * 1.2
decreasing_contrast_matrix = np.ones(img.shape) * 0.8

lower_contrast_image = np.uint8(cv2.multiply(np.float64(img), decreasing_contrast_matrix))
higher_contrast_image = np.uint8(np.clip(cv2.multiply(np.float64(img), increasing_contrast_matrix), 0, 255))

plt.figure(figsize = [18,5])
plt.subplot(131); plt.imshow(lower_contrast_image[:, :, ::-1]);  plt.title('Lower Contrast')
plt.subplot(132); plt.imshow(img[:, :, ::-1]);        plt.title('Original')
plt.subplot(133); plt.imshow(higher_contrast_image[:, :, ::-1]); plt.title('Higher Contrast');
```
